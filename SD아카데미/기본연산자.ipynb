```
score = int(input('What is your score'))
if score >= 90:
  letter = 'A'
elif score >= 80:
  letter = 'B'
elif score >= 70:
  letter = 'C'
elif score >= 60:
  letter = 'D'
else:
  letter = 'F'
print('your score is ' + letter)
```

```
x = 1
n = 0
# 변수 가지고 조건 만듬
# len(list) 를 주로 조건으로 씀
while x < len(6):
  score = int(input('What is your score\n'))
  x = x + 1
  if score >= 90:
    letter = 'A'
  elif score >= 80:
    letter = 'B'
  elif score >= 70:
    letter = 'C'
  elif score >= 60:
    letter = 'D'
  else:
    letter = 'F'
  n = n + 1
  print(n, 'your score is ' + letter, '\n')
```
