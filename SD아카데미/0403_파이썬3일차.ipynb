```
for num in range(10,20):
  for i in range(2, num):
    if num%i == 0:
      j = num/i
      print('%d equals %d * %d' %(num, i, j))
      break
  else:
    print(num, 'is a prime number')
```

```
seq1 = ['foo', 'bar', 'baz']
seq2 = ['one', 'two', 'three']
seq3 = ['apple', 'raspberry', 'banana']
```

```
for a, b, c in zip(seq1, seq2, seq3):
  print('a : {0}, b : {1}, c : {2}'.format(a, b, c))
```
