```
noun_list = []
verb_list = []
else_list = []

v = 0
c = 0
x = 0

count = 0
```

```
while count < len(split_list_1):

  if '다' in split_list_1[count]:
    verb_list.insert(v, split_list_1[count])
    v = v + 1

  elif '을'in split_list_1[count] or '를'in split_list_1[count] or '였던' in split_list_1[count]or '른'in split_list_1[count] or '했고' in split_list_1[count] or '게' in split_list_1[count] or '에' in split_list_1[count] or '은' in split_list_1[count]or '는데' in split_list_1[count]:
    else_list.insert(c, split_list_1[count])
    c = c + 1

  else:
    noun_list.insert(x, split_list_1[count].replace('[', '').replace('!', '').replace('=', ' ').replace(']', '').replace('는', '').replace('와', ''))
    x = x + 1
  

  count = count + 1
```
