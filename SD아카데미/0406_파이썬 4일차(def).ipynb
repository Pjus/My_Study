```
def input_sum():
  x = 1
  while x:
    first, second = input('\nInput two Number ?\n').split(',')
    result = int(first) + int(second)
    print('Input number :' + str(first) + ',' + str(second))
    print('Inside the function :', result)

    if result > 50:
      x = 0
  return result
```

```
total = input_sum()
print('outside fuction :', total)
```

```
def input_subtract():
  x = 1
  while x:
    first, second = input('Input two Number?\n').split(',')
    result = int(first) - int(second)
    print('Input number :' + str(first) + ',' + str(second))
    print('Inside the function :', result)

    if result > 50:
      x = 0
  return result
```

```
input_subtract()
```

```
def print_return_tuple(arg1, *arg2):
  print(arg1)
  for i in arg2:
    print(i)
  return arg2
```

```
print_return_tuple(10)
```

```
out_tuple = print_return_tuple(70,60,50)
```

```
def differenceVars(*var1, **var2):
  print(var1, type(var1))
  print(var2, type(var2))
  for i in var2:
    print(i)
```

```
differenceVars(1,2,3,4,5,6, a=2,b=3,c=4)
```
