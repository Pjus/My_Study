```
students_list = [
 [1, 90, 80, 85, 0, 0.0, 0],
 [2, 90, 85, 90, 0, 0.0, 0],
 [3, 80, 80, 80, 0, 0.0, 0]
]
```

```
students_dict = [
{'num':'1', 'name':'김철수','kor':90, 'eng':80, 'math':85},
{'num':'2', 'name':'박제동','kor':90, 'eng':85, 'math':90},
{'num':'3', 'name':'홍길동','kor':80,' eng':80, 'math':80}
]
```

```
students_list, students_dict
```

# indexing

```
students_list[0], students_dict[0]
```

```
students_list[0][1], students_dict[0]['name']
```

# slicing

```
students_list[:3], students_dict[:-1]
```

# update

```
students_list[1], students_dict[2]
```

```
# row
students_list[1] = [1, 95, 89, 88, 0, 0.0, 0]
students_dict[2] = {'num':'1', 'name':'김철수','kor':96, 'eng':83, 'math':92}
students_list[1], students_dict[2]
```

```
# column
students_list[1][2] = 92
students_dict[2]['eng'] = 94
students_list[1], students_dict[2]
```

## delete

```
students_list, students_dict
```

```
# row
del students_list[1]
del students_dict[2]
students_list, students_dict
```

```
# column
del students_list[0][0]
del students_dict[1]['name']
students_list, students_dict, len(students_dict)
```

# insert & append

```
students_list.insert(0, [1, 90, 80, 85, 0, 0.0, 0])
students_dict.insert(len(students_dict), {'eng': 94, 'kor': 96, 'math': 92, 'name': '김철수', 'num': '4'})
students_list, students_dict
```

```
students_list[0].insert(0, 30)
students_dict[2]['temp']= 30
students_list, students_dict
```
