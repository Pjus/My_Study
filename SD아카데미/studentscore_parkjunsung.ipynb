```
# 기본정보

students_score = [
                  {'num' : '1', 'name' : '김철수', 'kor' : 90, 'eng' : 80, 'math' : 85},
                  {'num' : '2', 'name' : '박제동', 'kor' : 85, 'eng' : 90, 'math' : 75},
                  {'num' : '3', 'name' : '홍길동', 'kor' : 95, 'eng' : 85, 'math' : 90},

]

  
```

```
# 추가
students_score.append({'num' : '4', 'name' : '김경섭', 'kor' : 88, 'eng' : 85, 'math' : 90})
students_score.append({'num' : '5', 'name' : '손진수', 'kor' : 96, 'eng' : 85, 'math' : 86})
```

```
students_score
```

```
# 과목별로 점수 따로 저장 학급평균
kor = []
eng = []
math = []
class_avg = []

# 학생별 과목 더해서 총합, 평균 추가 
for num, student in enumerate(students_score):
  student['total'] = student['kor'] + student['eng'] + student['math']
  student['avg'] = round((student['kor'] + student['eng'] + student['math']) / 3)
  
  class_avg.insert(num, student['avg'])

  # 리스트에 과목별 점수 추가
  kor.insert(num,student['kor'])
  eng.insert(num, student['eng'])
  math.insert(num, student['math'])

```

```
# 과목별 평균 구함
kor_avg = sum(kor) / len(students_score)
eng_avg = sum(eng) / len(students_score)
math_avg = sum(math) / len(students_score)

# 학급평균 구함
class_avg_score = sum(class_avg) / len(students_score)

# 출력
print('class_avg : {}'.format(class_avg_score))
print('kor_avg : {0}\neng_avg : {1}\nmath_avg : {2}'.format(kor_avg, eng_avg, math_avg) )
```

```
# 학생별 평균, 총점 출력
for i in students_score:
  print(i['name'] + '_avg ' + ":", i['avg'], 'total :', i['total'])
```
