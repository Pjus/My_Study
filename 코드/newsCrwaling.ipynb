#증권 뉴스 크롤링

```
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
```

## 날짜 범위 지정

```
date = '2020-04-21'
dt_index = pd.date_range(start='20100101', end='20200420')
```

## TEST

### 뉴스 url 가져오기

```
base = 'https://finance.naver.com/'
url = 'https://finance.naver.com/news/mainnews.nhn?date=' + date
```

```
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
```

```
url_list = []

for i in soup.find('ul', class_='newsList'):
  if type((i.find('a'))) != int:
    print(i.find('a')['href'])
    news_url = base + i.find('a')['href']
    url_list.append(news_url) 
```

```
url_list
```

### 뉴스 본문 크롤링

```
news_list = []
for i in url_list:
  time.sleep(1)
  res = requests.get(i)
  soup = BeautifulSoup(res.content, 'html.parser')
  content = soup.find('div', id='content')
  news_list.append(content.text.strip())
```

```
for i in news_list:
  print(i.strip())
```

## 날짜별 뉴스 크롤링 

### 각 뉴스 url 가져오기

```
dt_index[3755:-1]
```

```
url_list = []

# 2020-04-13 부터 2020-04-19 까지
for i in dt_index[3755:-1]:
  date = str(i)[:10]
  base = 'https://finance.naver.com/'
  url = 'https://finance.naver.com/news/mainnews.nhn?date=' + date

  time.sleep(2)

  res = requests.get(url)
  soup = BeautifulSoup(res.content, 'html.parser')
  

  for j in soup.find('ul', class_='newsList'):
    if type((j.find('a'))) != int:
      # print(j.find('a')['href'])
      news_url = base + j.find('a')['href']
      url_list.append(date +"/"+ news_url) 
```

```
len(url_list)
```

```
news_list = []
for i in url_list:
  time.sleep(2)
  res = requests.get(i[11:])
  soup = BeautifulSoup(res.content, 'html.parser')
  content = soup.find('div', id='content')
  news_list.append(i[:11] + content.text.strip())
```

```
news_list
```

### 날짜 / 뉴스 본문 분리해서 데이터 프레임으로 만들기

```
Date = []
News = []

for i in news_list:
  date = i[:10]
  news = i[11:]
  Date.append(date)
  News.append(news)
```

```
df_date = pd.DataFrame(Date)
df_news = pd.DataFrame(News)
```

```
df_all = pd.concat([df_date, df_news], axis=1)
```
