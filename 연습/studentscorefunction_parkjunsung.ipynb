```
# 기본정보

students_score = [
                  {'num' : '1', 'name' : '김철수', 'kor' : 90, 'eng' : 80, 'math' : 85},
                  {'num' : '2', 'name' : '박제동', 'kor' : 85, 'eng' : 90, 'math' : 75},
                  {'num' : '3', 'name' : '홍길동', 'kor' : 95, 'eng' : 85, 'math' : 90},

]
```

```
# 추가
students_score.append({'num' : '4', 'name' : '김경섭', 'kor' : 88, 'eng' : 85, 'math' : 90})
students_score.append({'num' : '5', 'name' : '손진수', 'kor' : 96, 'eng' : 85, 'math' : 86})
```

```
# total avg 추가 함수
def total_avg(students_score):

  for student in (students_score):
    student['total'] = student['kor'] + student['eng'] + student['math']
    student['avg'] = round((student['kor'] + student['eng'] + student['math']) / 3)

    print('이름 :', student['name'])
    print('총합 :', student['total'])
    print('평균 :', student['avg'])
    print('\n')

  return students_score
```

```
students_score = total_avg(students_score)
```

```
def class_avg(students_score):

  class_avg = 0

  for student in (students_score):
    class_avg += student['avg']

  print('학급평균 :', class_avg/len(students_score))

  return class_avg/len(students_score)
```

```
class_avg = class_avg(students_score)
```

```
def sub_avg(students_score):

  kor = 0
  eng = 0
  math = 0

  for student in (students_score):
    kor += student['kor']
    eng += student['eng']
    math += student['math']

  print('국어평균 :', kor/len(students_score))
  print('영어평균 :', eng/len(students_score))
  print('수학평균 :', math/len(students_score))

  return kor/len(students_score), eng/len(students_score), math/len(students_score)
```

```
kor_avg, eng_avg, math_avg = sub_avg(students_score)
```
