```
from matplotlib import pyplot as plt 
import numpy as np
```

```
student_score = [
[1, 90, 80, 85, 0, 0.0, 0, 0],
[2, 88, 85, 90, 0, 0.0, 0, 0],
[3, 32, 85, 90, 0, 0.0, 0, 0],
[4, 42, 75, 93, 0, 0.0, 0, 0],
[5, 82, 86, 88, 0, 0.0, 0, 0]
]
```

```
student_score = np.array(student_score)
```

```
x = ['student1', 'student2', 'student3', 'student4', 'student5']
c = ['kor', 'eng', 'math']
n = 0
for i in range(1,4):
  y = student_score[:, i]
  plt.plot(x, y, 'o-', label = c[n])
  n = n +1
  
plt.legend()
plt.show()
```

```
x = ['kor','eng','math']
for i in range(len(student_score)):
  y = student_score[i, 1:4]
  plt.plot(x, y, 'o--')
  
plt.legend(['student1', 'student2', 'student3', 'student4', 'student5'])
plt.show()
```

```
x1 = [1, 3, 4, 5, 6, 7, 9] 
y1 = [4, 7, 2, 4, 7, 8, 3]
x2 = [2, 4, 6, 8, 10]
y2 = [5, 6, 2, 6, 2]
plt.bar(x1, y1, label="Blue Bar", color='b')
plt.bar(x2, y2, label="Green Bar", color='g')
plt.xlabel("bar number")
plt.ylabel("bar height")
plt.title("Bar Chart Example")
plt.show()
```

```
labels = ['s1', 's2', 's3', 's4', 's5']

x = np.arange(len(labels))  # the label locations
width = 0.15  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, student_score[:, 1], width, label='kor')
rects2 = ax.bar(x + width/2, student_score[:, 2], width, label='eng')
rects3 = ax.bar(x + 3*(width/2), student_score[:, 3], width, label='math')

ax.set_ylabel('Scores')
ax.set_title('Scores by group')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```

```
n = 5 + np.random.randn(1000)
m = [m for m in range(len(n))]

plt.bar(m,n)
plt.title('Raw Data')
plt.show()
```

```
plt.hist(n, bins=20)
plt.title('Histogram')
plt.show()
```

```
plt.hist(n, cumulative=True, bins = 20)
plt.title('Cumulative Histogram')
plt.show()
```

```
plt.hist(n, cumulative=False, bins = 20)
plt.title('Cumulative Histogram')
plt.show()
```
