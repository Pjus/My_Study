```
from matplotlib import pyplot as plt 
import numpy as np
```

```
x = np.arange(1,11)
y = 2*x + 5
```

```
plt.title('Matplotlib Demo')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x,y, 'bo')
plt.show()
```

```
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1]
y2 = [2, 4, 6, 4, 2, 4, 6, 4, 2]
plt.plot(x, y1, 'bo--', label="line L")
plt.plot(x, y2, 'ro--', label="line H")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Graph Example")
plt.show()
```

```

student_score = [
[1, 90, 80, 85, 0, 0.0, 0, 0],
[2, 88, 85, 90, 0, 0.0, 0, 0],
[3, 62, 85, 90, 0, 0.0, 0, 0],
[4, 80, 75, 93, 0, 0.0, 0, 0],
[5, 82, 86, 88, 0, 0.0, 0, 0]
]
```

```
student_score = np.array(student_score)
```

```
student_score[:, 1]
```

```
x = ['1', '2', '3', '4', '5']
plt.plot(x, student_score[:, 1], 'bo--', label="line L")
plt.plot(x, student_score[:, 2], 'ro--', label="line L")
plt.plot(x, student_score[:, 3], 'go--', label="line L")

plt.show()
```

```
x = ['kor','eng','math']
for i in range(len(student_score)):
  y = student_score[i, 1:4]
  plt.plot(x, y, 'o--', label="line L")
plt.show()
```
