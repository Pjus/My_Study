```
import numpy as np
from collections import Counter as cc
```

```
student_scores_high = np.random.randint(low = 80, high=100, size =(30, 3)) 
student_scores_middle = np.random.randint(low = 50, high=79, size =(20, 3)) 
student_scores_low = np.random.randint(low = 0, high=49, size =(100, 3)) 
student_scores_high
```

```
student_scores = np.insert(student_scores_high, 0, student_scores_middle, axis=0)
student_scores = np.insert(student_scores, 0, student_scores_low, axis=0)
```

```
student_scores
```

```
len(student_scores)
```

```
sub = ['수학', '영어', '국어']
matt = np.zeros([3,4])
for i in range(len(sub)):
  
  print(sub[i], '\n')

  x = student_scores[:, i]
  
  x_min = x.min()
  x_max = x.max()
  x_mean = x.mean(dtype = np.int64)
  x_var = x.var(dtype = np.int64)
  x_std = x.std(dtype = np.int64)
  x_meadian = np.median(x)
  x_cnt = cc(x) # 데이터에서 가장 많이 나온 값 카운트
  x_cntmin = x_cnt.most_common(1)

  print('평균 :', x_mean)
  print('분산 :', x_var)
  print('중앙값 :', x_meadian)
  print('최소값 :', x_min)
  print('최대값 :', x_max)
  print('최빈값 :', x_cntmin[0])
  print('표준편차 :', x_std)
  print('\n')
  mat = np.array([[x_mean, x_meadian, x_std, x_var]])
  matt[i] = mat

# 수학, 영어, 국어 대표값 : 중앙값
# 선택이유 : 자료 중에 극단적인 값이 한 개 이상 있을 때 그 값들이 산술평균에 큰 영향을 미쳐서 산술평균이 대표값으로 부적절한 경우가 있다. 
# 이런 경우에는 극단적인 값의 영향을 받지 않는 중위수나 최빈수가 오히려 대표값으로 적절할 수 있다.
```

```
n = 0
print('\t평균', '중앙값', '표준편차', '분산')
for i in matt:
  print(sub[n], i)
  n = n + 1
```
