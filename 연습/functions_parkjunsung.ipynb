```
def sum(first, second):
  result = int(first) + int(second)

  print('Input number :' + str(first) + ',' + str(second))
  print('Inside the function :', result)

  return result
```

```
first, second = input('Input two Number ?\n').split(',')
```

```
total = sum(first, second)
print('outside fuction :', total)
```

```
def input_sum():
  x = 1
  while x:
    first, second = input('\nInput two Number ?\n').split('+')
    result = int(first) + int(second)
    print('\nInput number :' + str(first) + ',' + str(second))
    print('Inside the function :', result)

    if result > 50:
      x = 0
      print('Exit when result > 50')
  return result
```

```
def input_subtract():
  x = 1
  while x:
    first, second = input('\nInput two Number ?\n').split('-')
    result = int(first) - int(second)
    print('\nInput number :' + str(first) + ',' + str(second))
    print('Inside the function :', result)

    if result < 0:
      x = 0
      print('Exit when result < 0')
  return result
```

```
def input_multi():
  x = 1
  while x:
    first, second = input('\nInput two Number ?\n').split('*')
    result = int(first) * int(second)
    print('\nInput number :' + str(first) + ',' + str(second))
    print('Inside the function :', result)

    if result > 1000:
      x = 0
      print('Exit when result > 1000')

  return result
```

```
def input_div():
  x = 1
  while x:
    first, second = input('\nInput two Number ?\n').split('/')
    result = int(first) / int(second)
    print('\nInput number :' + str(first) + ',' + str(second))
    print('Inside the function :', result)

    if result < 0:
      x = 0
      print('Exit when result < 0')

      
  return result
```

```
total = input_sum()
print('outside fuction :', total)
```

```
subtract = input_subtract()
print('outside fuction :', subtract)
```

```
multi = input_multi()
print('outside fuction :', multi)
```

```
div = input_div()
print('outside fuction :', div)
```
