#더 해보기

```
# 이데일리 메인 기사 & 링크

url = 'https://www.edaily.co.kr/'

res = requests.get(url)

soup = BeautifulSoup(res.content, 'html.parser')

c = soup.find_all('section', class_='double_bg')
```

```
len(c)
```

```
c[0]
```

```
for news in c[0].find_all('div', class_='img_subtitle'):
  print(news.find('a')['href'])
  print(news.find('a')['title'])
  print('\n')
```

```
for news in c[1].find_all('li', class_='trend_1'):
  print(news.find('a')['href'])
  print(news.find('a')['title'])
  print('\n')
```

```
# sbs 메인 기사 & 링크

url = 'https://news.sbs.co.kr/news/newsMain.do?div=pc_news'

res = requests.get(url)

soup = BeautifulSoup(res.content, 'html.parser')

v = soup.find_all('div', class_ ='w_inner')
```

```
len(v)
```

```
type(v), len(v)
```

```
for i in v:
  print(type(i))
```

```
for i in v:
  for j in i.find_all('li'):
    for k in j.find_all('p', class_='desc'):
      print(k.text)
      print(j.find('a')['href'])
```
